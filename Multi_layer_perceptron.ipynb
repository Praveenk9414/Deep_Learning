So lets first see how it captures the non-linearity in the data -> the linear output from each of the initial perceptron are combined and then smoothen.

How the smoothening is done -> 

    Firstly lets see how the output is caluclated the prob from both the inital perceptron is again computed using (w1 * x1 + w2 * x2 + b) and then it is again passed through an activation function (sigmoid) -> which brings it in the range (0-1).
