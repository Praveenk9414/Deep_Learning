```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer = Tokenizer()
```

```python
tokenizer.fit_on_texts([faqs])   # providing a list b/c we can provide multiple inputs
```

```python
tokenizer.word_index          # dictionary to see the int value assigned to each unqiue words
```

```python
input_sequences = []
for scentence in faqs.split('\n'):
    tokenized_sequence = tokenizer.texts_to_sequences([scentence])[0]     # this one converts the text into sequence

    for i in range(1,len(tokenized_sequence)):
        input_sequences.append(tokenized_sequence[:i+1])
input_sequences
```

```python
maxlen = max([len(i) for i in input_sequences])
```

```python

from tensorflow.keras.utils import pad_sequences
padded_input_sequences = pad_sequences(input_sequences, maxlen=maxlen, padding='pre')
```

```python
padded_input_sequences
```

```python
# splitting the data into input and output
X = padded_input_sequences[:,:-1]    
y = padded_input_sequences[:,-1]
```

Now as we are approaching this problem as a supervised ML problem 

So we need to know- what type of supervised model -> regression (or) classification 

Regression output (continuous) -> will be like 3.4, 2.1 5.6 -> but we have nothing defined as this 2.1 or 5.6 so this will show an error ... so obviously it will be an classification problem that too a multiclass classification prb.

For which we will use One hot encoding -> [0001000, 01000000, ...] this will be the output structure

```python
# this tells us the number of classes
len(tokenizer.word_index)
```

```python
from tensorflow.keras.utils import to_categorical

# why num_classes = 283 -> As it starts with 1, and so that it doesnot miss the last one
y = to_categorical(y, num_classes=283)
y.shape
```

```python
y
```

![lec_img/lstm_wordpred_arch.png](lec_img/lstm_wordpred_arch.png)

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
```

```python
model = Sequential()

# 283-> Input_dim(no of words), 100-> Output_dim(hyperparamter), Input_length-> length of input sequence after padding
model.add(Embedding(283, 100, input_length=56))
model.add(LSTM(150))
model.add(Dense(283, activation='softmax'))
model.build(input_shape=(None, 56))  # Explicitly build
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.summary()
```

Now lets understand how the input will be send ->

1. Input - lets say the first line -> (1,56) -> 56 numeric repr of words with padding
2. Now each word from the 56 words is represented with 100 numeric values using embedding layer -> giving 28300 parameters total.(Here we are converting the inital words into dense vector).
3. And then each of the words(100 vector repr) is given to the LSTM based on the timestamp -> t1, t2, t3, .... t56.
4. And since the no of nodes of lstm is 150 the output from here would be (1,150) for each words.
5. Futher we are applying a dense layer of 283 nodes for multi-class classification.
6. Now each of these will have softmax as output and will provide a probability for each class of 283 i.e b/w (0-1) { e.g. 0.3, 0.5, 0.67... }, and the one with the max prob will be the output

```python
history = model.fit(X,y,epochs=100)
```

```python
# now lets do some predictions 
import numpy as np

text = "i assume"
for i in range(5):
    #tokenize
    token_text = tokenizer.texts_to_sequences([text])[0]
    
    # padding
    padded_token_text = pad_sequences([token_text], maxlen=56, padding='pre')
    
    # predict
    pos = np.argmax(model.predict(padded_token_text))
    for word,index in tokenizer.word_index.items():
        if index == pos:
            text = text + " " + word
            print(text)
```
