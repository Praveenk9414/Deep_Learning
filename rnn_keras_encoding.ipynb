To apply RNN on textual data we need to apply numeric conversion so that we can apply RNN on it...

Keras has two apporach to handle text to numbers conversion

    - Integer Encodig
    - Embedding


```python
# Integer encoding SimpleRNN

import numpy as np

docs = ['go india',
    'india india',
    'hip hip hurray',
    'jeetega bhai jeetega india jeetega',
    'bharat mata ki jai',
    'kohli kohli',
    'sachin sachin',
    'dhoni dhoni',
    'modi ji ki jai',
    'inquilab zindabad']
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
tokenizer = Tokenizer(oov_token='<nothing>')
```

What this do is create a {word : index} dictionary

oov_token -> Out-Of-Vocabulary Token -> it assign any other new word to 'xyz'

Text to sequence -> Example: "I love Python" → [1, 2, 4]

Count word Frequency -> tokenizer.word_counts

```python
tokenizer.fit_on_texts(docs)
```

```python
tokenizer.word_index
```

```python
tokenizer.word_counts
```

```python
tokenizer.document_count
```

```python
sequences = tokenizer.texts_to_sequences(docs)
sequences
```

```python
from tensorflow.keras.utils import pad_sequences
```

```python
sequences = pad_sequences(sequences, padding='post') # 'post' -> the zero padding will be adding to the end
sequences
```

```python
# Now lets apply it on real dataset

from tensorflow.keras.datasets import imdb
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, SimpleRNN, Embedding, Flatten
```

```python
(X_train, y_train),(X_test, y_test) = imdb.load_data()
```

```python
X_train.shape
```

Since the length of each of the elements within the X_train & X_test are diff so only the no of elements are show ...

```python
X_test.shape
```

```python
# keeping only 50 words of each scentence to reduce computation
X_train = pad_sequences(X_train, padding='post', maxlen=50)
X_test = pad_sequences(X_test, padding='post', maxlen=50)
```

```python
X_train.shape
```

```python
model = Sequential()

model.add(SimpleRNN(32, input_shape=(50,1), return_sequences=False))
model.add(Dense(1, activation='sigmoid'))

model.summary()
```

Why keep the return_sequences -> False

As we are perfoming Sentiment analysis so their is no need to output at each step rather we only need the output from the last step.

Named Entity Recognition, Machine Translation -> these are some of the process where we need the output at each step.

![lec_img/rnn_map.png](lec_img/rnn_map.png)

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

model.fit(X_train, y_train, epochs=5, validation_data=(X_test, y_test))
```

## Embedding

In NLP, word embedding is a term used for representation of the words for text analysis. It is a way of representing words as dense vectors of real numbers instead of treating them as one-hot vectors or simple integers.

    - One-hot encoding: “cat” → [0, 0, 1, 0, 0]
        (very sparse, no meaning captured, large dimensional space)
    - Word embedding: “cat” → [0.25, -0.61, 0.11, 0.89, …]
        (dense, low-dimensional, captures semantic meaning)
So, words that have similar meanings end up with vectors that are close to each other in this embedding space.

Example:

    - “king” and “queen” → vectors are close.
    - “cat” and “dog” → vectors are closer than “cat” and “car”.

In Deep Learning -> keras provide a feature to add embedding layer

```python
# Using the same sequence (which we got from padding)
sequences
```

```python
# adding a embedding layer on it

model = Sequential()

# it takes 3 inputs ... 
#        1. input_len -> Size of vocabulary (maximum integer index + 1); in this case 18 (17+1)
#        2. output_dim -> with how many numbers do you want to represent your words (hyperparameter) -> you tune it and find the best fit
#        3. input_length -> Length of input sequences after padding
model.add(Embedding(input_dim=18, output_dim=2, input_length=5))
model.summary()
```

```python
model.compile('adam', 'accuracy')
```

```python
# now this is the vector reprsentation of each sequences, where each number is represented by 2 numbers
pred = model.predict(sequences)
pred
```
