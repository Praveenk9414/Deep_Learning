### DeepRNN -
Recurrent Neural Network with multiple hidden layers stacked on top of each other (like a deep feedforward net but in the recurrent world).

Architecture of DeepRNN

Here the stack above is the hidden layer ...

- Here *orange circle* shows the recurrent data transfered from the the (t-1)s timestamp (the first nodes layer) to the (t)s timestamp (the first nodes layers)....

- The  *blue circle*   shows the recurrent data transfered from the the (t-1)s timestamp (the second nodes layer) to the (t)s timestamp (the second nodes layers)....

- The  *Green circle*  shows the forward data transfer in the same timestamp...

Why to use DeepRNN ->

    1. Hierarchical Repr - 
![lec_img/drnn_stack.png](lec_img/drnn_stack.png)

    The benefit of structure is that it can capture every details from the previous word or scentence.
    For e.g  1. "The audio is bad"       -> At the premetive layer they will try to understand at the word level
             2. "Display is great"       -> understand at the scentence level
             3. "But overall i am happy" -> and then understand it based on the whole paragraph

        Hence understanding the sentiment of the paragraph at each level.

    By creating a Hierarchical appraoch to understand the data.

    2. Customization for advance task -
        In an attention mechanism you would use deeprnn for the encoder & decoder 
        for e.g -> google translate uses deeprnn to encode & decode the paragraphs

    

When to use Deep RNN

![lec_img/drnn_when.png](lec_img/drnn_when.png)

1. You have to perform complex tasks
2. When you have large datasets (small dataset can lead to overfitting in deeprnn)
3. If you have sufficient computational resources
4. Used as Improvement - if simple model doesn't shows significant output then we use Deeprnn.

```python
#Now lets implement Deep RNN in keras

import tensorflow as tf
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, Dense, LSTM, GRU
```

```python
#imdb dataset

(X_train, y_train) , (X_test, y_test) = imdb.load_data(num_words=10000)

# pad_sequences to have the same length
X_train = pad_sequences(X_train, maxlen=100)
X_test = pad_sequences(X_test, maxlen=100)
```

```python
# defining the RNN model

model = Sequential([
    Embedding(10000, 32, input_length=100),     # taking 10000 words , repr each word with 32 params of vector
    SimpleRNN(5, return_sequences=True),        # RNN layers with 5 nodes 
    SimpleRNN(5),                               # Another RNN layers with 5 nodes 
    Dense(1, activation='sigmoid')
])

# return sequence are kept true for all the inner hidden nodes ... the last one is kept default false
```

![lec_img/deeprnn_code.png](lec_img/deeprnn_code.png)

```python
model.build(input_shape=(None, 100))
```

```python
model.summary()
```

### This variation of SimpleRNN to create stack can be used for LSTM and GRU as well
In the contrary DeepLSTM and DeepGRU are more used compared to DeepRNN

Having the very same implementation ...

```python
model = Sequential([
    Embedding(10000, 32, input_length=100),
    LSTM(5, return_sequences=True),
    LSTM(5),
    Dense(1, activation='sigmoid')
])
model.build(input_shape=(None, 100))
model.summary()
```

```python
model = Sequential([
    Embedding(10000, 32, input_length=100),
    GRU(5, return_sequences=True),
    GRU(5),
    Dense(1, activation='sigmoid')
])
model.build(input_shape=(None, 100))
model.summary()
```
