### 1. In Many to One ->
Muliple inputs at different timestamp ... whereas we recieve only one single output at the end.

![lec_img/many2one.png](lec_img/many2one.png)

### 2. In One to Many ->

we have a non-sequential input(like an img, audio) -> and generate mulitple outputs

e.g; image captioning(describing the image in scentenc), music generation(generating multiple tunes)

![lec_img/one2many.png](lec_img/one2many.png)

### 3. In Many to Many (seq 2 seq)

I/P & O/P -> both are sequential

Two types ->

    - Same length (len(I/P seq) == len(O/P seq))
    - Variable length 

1. Same length ->
   e.g;
   - POS (parts of speech) tagging -> u assign each with a grammer tag
   - NER (Named Entity Recognition) -> to check wether it is (or) not an entity -> binary (assign 1 0 0 1 0 0 1) ->like that

![lec_img/many2many.png](lec_img/many2many.png)

### IMP
2. Variable length ->e.g;
   - Machine Translation -> google translater converting english into hindi(the number words might vary)
     
In this Variable length what happens is firstly every input is read and then after the last input the output generation is started. Hence making a encoder - decoder structure.

Why -> so that it can understand the grammatical meaning & context of the scentence

![lec_img/var_len.png](lec_img/var_len.png)

### 3. One to One

Here both I/P & O/P are non-sequential
e.g; Image classification (I/P is an image -> O/P is binary classication cat or not)

So technically this is a simple Neural Network -> i.e what we have been dealing with earlier ANN & CNN
