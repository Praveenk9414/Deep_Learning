## Deep Learning V/S Machine Learning

1. **Data Dependency** - Deep Learning models are data hungry -> the more amt of data you give -> the better the model performes.
![](lec_img/data_hungry.png)

The graph shows how the ML models perfromance after a certain points stagnates(that is no significant imporvement) ... whereas in the DL the Performance inc as the data inc.

2. **Hardware Dependency** - Deep learning models do a lot matrix multi inside -> so require GPU's for faster calculations. (Costly hardware)
3. **Training Time** - Since the Deep Learning models are complex - so they may require weeks to months sometime ... whereas ML models takes less time comparitively.
   While the predictions times various in both the cases -> Predictions in DL are generally faster ... whereas predictions in KNN can be more time consuming.
4. **Feature Selection** - We already saw -> feature learning (repr learning) -> DL (extracts the features itself directly from the raw data) ... whereas in ML(features need to be made by - domain experts)
5. **Interpretability** - Since the features are extracted automatically -> we can't interpret the answer or say give a reason for why that answer ... The models in DL are blackbox -> so creates a interpretability issue ...
   Whereas the interpretability in ML is very high -> we always know the features and can reason why a specific response.


So depending on the need and applicability -> we use ML or DL -> both having very much significance.
