## Problem with RNN (or) The need for LSTM ->

Traditional RNNs have difficulty learning long-term dependencies due to the vanishing and exploding gradient problems. 

This makes them less effective for tasks where context over long sequences is important. 

LSTM (Long Short-Term Memory) networks were developed to address these issues. They have special gating mechanisms that help them retain important information for longer periods and prevent the gradients from vanishing or exploding.
