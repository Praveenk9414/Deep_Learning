### Loss functions v/s Cost functions - 

Loss func - Is the error calculated for a single row of the data...

Whereas,
Cost func - Is the sum of all the error calculated throughout the datasets ...

### Loss functions -

1. For Regression -

   - **MSE (Mean Squared Error)** -> 
         ![](lec_img/mse.png)
  
     Advantage -
     
          - Easy to Interpret
          - Differentiable curve (Upwards parabolla)
          - 1 Local Minima
  
     Disadvantage -

          - Error unit is the square of the Actual thing
          - Not Robust to outliers (Since the error is squared -> outliers could cause major changes in the predictions slope)

    [IMP] You can only apply MSE when you Output activation function is -> linear func
   
   ![](lec_img/mse_after_linear.png)

    $$ Cost function (with MSE) = \frac{1}{m} \sum_{i=1}^{m} \left( y^{(i)} - \hat{y}^{(i)} \right)^2 $$
    
   - **MAE (Mean Absolute Error) / L1 Loss** ->

     $$
Loss = \left| y - \hat{y} \right|
$$
         $$
Cost (with MAE) = \frac{1}{m} \sum_{i=1}^{m} \left| y^{(i)} - \hat{y}^{(i)} \right|
$$

        Advantage -
       
            - Intuitive & Easy
            - Unit same as the actual thing
            - Robust to Outliers
        Disadvantage -
    
            - Not differntiable
            - Since not diffentiable -> we can't rely on Gradient descent -> we have to use Sub-Gradients -> which increases the complexity


    - **Huber Loss** ->
  
          Huber Loss is basically the combination of both the MSE & MAE.
  
          - We know that MSE works poor in the case of outliers, but if not an outlier works good on normal points.
          - Whereas MAE works good for outliers.
  
      Why even need this ... if to many outliers then they are not outliers -> they are to be termed as part of the data -> so to consider those points and have a better pred -> we use Huber loss.
  
![](lec_img/huber_loss.png)
  
Where small delta -> is the hyperparameter -> which define whether a point outlier or not (could be adjusted)
