```python
from keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
img = image.load_img('train/dog.jpeg', target_size=(200,200))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(img)
```

```python
type(img) # PIL -> python image library object
```

```python
img = image.img_to_array(img)
img.shape
```

```python
input_batch = img.reshape(1,200,200,3)
```

```python
datagen = ImageDataGenerator(
        rotation_range=40,
        width_shift_range=0.2,
        height_shift_range=0.2,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True,
        fill_mode='nearest')
```

```python
i = 0
for batch in datagen.flow(input_batch, batch_size=1,
                          save_to_dir='train/train_set', save_prefix='dog', save_format='jpeg'):
    i += 1
    if i == 10:
        break  # otherwise the generator would loop indefinitely
```
